```python
model_id="Narrativaai/BioGPT-Large-finetuned-chatdoctor"
filenames = [
    "pytorch_model.bin",
    "config.json",
    "Training_args.bin",
]
```

```python
download_dir = "biomistral_model"
os.makedirs(download_dir, exist_ok=True)

model_id_str = "Narrativaai/BioGPT-Large-finetuned-chatdoctor"
model_paths = {}

for filename in filenames:
  model_paths[filename] =hf_hub_download(
      repo_id=model_id_str,
      filename=filename,
      cache_dir=download_dir,
      token=os.environ["HUGGING_FACE_API_KEY"],
  )
```

```python
model_paths = os.path.join(download_dir, "pytorch_model.bin")
if not os.path.exists(model_paths):
    raise ValueError(f"Model path does not exist: {model_paths}")
```

```python
llm = LlamaCpp(
    model_path =os.path.join(download_dir,"pytorch_model.bin"),
    n_ctx = 2048,
    verbose = True
    )
```

```python
template = """
<|context|>
You are a medical assistant that follows the instruction and generate the accurate response based on the query and the context provided. Give a direct answers.
</s>
<|user|>
{query}
</s>
<|assistant|>
"""
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import ChatPromptTemplate
from langchain.llms import LlamaCpp
from huggingface_hub import hf_hub_download
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
llm_chain = LLMChain(llm=llm, promt=promt)
```

```python
rag_chain = (
    {"context": retriever, "query": RunnablePassthrough()}
    |prompt
    |llm_chain
    |llm
    |StrOutputParser()
)
```

```python
response=rag_chain.invoke(query)
```
